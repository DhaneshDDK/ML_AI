```python
# class

class Car:
    def __init__(self, name, power,cost):
         self.name = name     #public access modifier
         self._power = power  #protected access modifier.  _  (variable - _name) -> same for methods
         self.__cost = cost   #private access modifier.    __ (variable - __name) -> ""
    
    def carPower(self,name):
         if(self.name == name):
              return self._power
         return "Car is not available"
    
    def findCost(self,name, power):
         if(self.name==name and self._power == power):
              return self.__cost
         return "Car is not available"

audi = Car("audi",50,1000)
print(audi)
print(audi.name)
# print(audi.__cost)
print(audi._power)
print(audi.carPower('audi'))
print(audi.carPower('bmw'))
print(audi.findCost('audi',50))
print(audi.findCost('audi',30))
```

```python
# Single Inheritance

class Car:
    def __init__(self, name, modal, windows):
        self.name = name
        self.modal = modal
        self.windows = windows
    
    def startCar(self, name, modal):
        if(self.name == name and self.modal == modal):
            print("Car engine is started, Ready to drive!!!")
        else:
            print("Trouble starting the car!")

class BMW(Car):
    def __init__(self,name,modal,windows,is_selfDriving):
        super().__init__(name,modal,windows)
        self.is_selfDriving = is_selfDriving

    def autoPilot(self,name,modal):
        if(self.name == name and self.modal == modal):
            print("Relax! You are auto pilot mode.")
        else:
            print('Error! Please drive carefully!!')

bmw = BMW('bmw',10,6,True)
bmw.autoPilot('bmw',10)

# multi-level inheritance
class Tesla(BMW):
    def __init__(self,name,modal,windows,is_selfDriving,is_electric):
        super().__init__(name,modal,windows,is_selfDriving)
        self.is_electric = is_electric
    
    def setBatteryMode(self,name,modal):
        if(self.name == name and self.modal == modal):
            print("Running on battery!")
        else:
            print("Electric mode is not available")

tesla = Tesla('tesla',10,5,True,True)
tesla.setBatteryMode('tesla',10)

# multiple inheritance
class Animal:
    def __init__(self,name):
        self.name = name

class Pet:
    def __init__(self,owner):
        self.owner = owner
    
class Dog(Animal,Pet):
    def __init__(self,name,owner):
        Animal.__init__(self,name)
        Pet.__init__(self,owner)
    
    def Bark(self,name,owner):
        if(self.name == name and self.owner == owner):
            print(f"{self.name} is barking boww boww!!, owned to {self.owner}")

lusi = Dog('lusi','losipher')
lusi.Bark('lusi','losipher')
```

```python
# Polymorphism

class Animal:
    def speak(self):
        print("making sound!!!")
class Dog(Animal):
    def speak(self):
        print("bow bow!!!")
class Lion(Animal):
    def speak(self):
        print("roar roar!!!")

def speak(animal):
    animal.speak()

dog = Dog()
lion = Lion()
speak(dog)
speak(lion)

#with Interface 

from abc import ABC,abstractmethod
class Vehicle(ABC):
    @abstractmethod 
    def start_engine(self):
        print("Engine is started...")
class Car(Vehicle):
    def start_engine(self):
        print("Car engine is started")
class Bike(Vehicle):
    def start_engine(self):
        print("Bike engine is started")
class Dummy(Vehicle):
    def start_engine(self):
        return super().start_engine()

def start_engine(vehicle):
    vehicle.start_engine()
car = Car()
bike = Bike()
dummy = Dummy()
start_engine(car)
start_engine(bike)
start_engine(dummy)
```

```python
# Encapsulation is the concept of wrapping data (variables) and methods(functions) together as a single unit. It restricts direct access to some
# of the object's components, which is a means of preventing accidental interface and misuse of the data

class Account:
    def __init__(self,name,accountNumber,balance,pincode):
        self.name = name
        self.accountNumber = accountNumber
        self._balance = balance
        self.__pincode = pincode
    
    def fetchPincode(self,name,accountNumber):
        if(self.name == name and self.accountNumber == accountNumber):
            return self.__pincode
        raise PermissionError("Access denied: Incorrect name or account number")

class Transaction(Account):
    def __init__(self, name, accountNumber, balance, pincode):
        super().__init__(name, accountNumber, balance, pincode)
        
    def debitAmount(self, amount, name, accountNumber,pincode):
        try:
            # Validate amount
            if not isinstance(amount, (int, float)) or amount <= 0:
                raise ValueError("Invalid amount. Please enter a positive number.")
            
            orgPincode = Account.fetchPincode(self,name,accountNumber)
            # First check pin
            if orgPincode != pincode:
                raise ValueError("Incorrect PIN")
            
            # Then check balance
            if self._balance < amount:
                raise ValueError("Insufficient funds in the account")
            
            # If all checks pass, perform the debit
            self._balance -= amount
            print(f"Amount debited: {amount}, Available balance: {self._balance}")
            
        except AttributeError:
            print("Error accessing account details")
        except ValueError as ex:
            print(str(ex))
        except Exception as ex:
            print(f"An unexpected error occurred: {str(ex)}")
        
    def creditAmount(self,amount,name, accountNumber,pincode):
        try:
            if not isinstance(amount,(int,float) or amount <=0):
                raise ValueError("Invalid amount, Please enter a positive number.")
            
            orgPincode = Account.fetchPincode(self,name,accountNumber)

            if orgPincode != pincode:
                raise ValueError("Incorrect PIN")
            
            self._balance += amount
            print(f"Amount created:{amount}, Available balance:{self._balance}")
        
        except Exception as ex:
            print('Error:', ex)
           


# Test the implementation
name = "Dhanesh"
accountNumber = 1234567890
balance = 10000
pincode = 1234

account = Account(name,accountNumber,balance,pincode)
transaction = Transaction(name,accountNumber,balance,pincode)

# Test different scenarios
transaction.debitAmount(10000, name, accountNumber, pincode)    # Should succeed
transaction.debitAmount(2000, name, accountNumber, 5678)        # Should fail - wrong PIN
transaction.debitAmount(20000, name, accountNumber,pincode)    # Should fail - insufficient funds
transaction.creditAmount(2000,name, accountNumber, pincode)    # Should succeed
transaction.creditAmount(0,name, accountNumber, pincode)   
transaction.creditAmount(0,name, accountNumber, 123)   
```

```python
# Example of Abstract Base Class with multiple abstract methods
from abc import ABC, abstractmethod

class Vehicle(ABC):
    @abstractmethod
    def start_engine(self):
        pass
    
    @abstractmethod
    def stop_engine(self):
        pass

class AudiCar(Vehicle):
    def start_engine(self):
        print("Audi engine started. rai rai rai...")
        
    def stop_engine(self):
        print("Audi engine stopped.")

class BMWCar(Vehicle):
    def start_engine(self):
        print("BMW engine started. vroom vroom...")
        
    def stop_engine(self):
        print("BMW engine stopped.")

class NanoCar(Vehicle):
    def start_engine(self):
        print("Nano engine started. put put put...")
        
    def stop_engine(self):
        print("Nano engine stopped.")

# Create instances
audi = AudiCar()
bmw = BMWCar()
nano = NanoCar()

# Test the implementations
audi.start_engine()
audi.stop_engine()

bmw.start_engine()
bmw.stop_engine()

nano.start_engine()
nano.stop_engine()
```

```python
## Magic methods

'''
Magic methods in Python, also known as dunder methods (double underscore methods), are special methods
that start and end with double underscores (__init__). These methods enable you to define the behaviour of objects 
for built-in operations, such as arithmetic operations, comparisions, and more
'''

''' 
__init__ -> initialize instance of a class
__str__  -> returns a string representation of object
__repr__ -> returns a official string representation of an object
__len__  -> return length of an object
__getitem__ -> gets an item from a container
__setitem__ -> sets an item in a container
'''

class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age
        self.hobbies = []

    def __str__(self):
        return f"Person(name={self.name}, age={self.age})"

    def __repr__(self):
        return f"Person('{self.name}', {self.age})"

    def __len__(self):
        return len(self.hobbies)

    def __getitem__(self, index):
        return self.hobbies[index]

    def __setitem__(self, index, value):
        self.hobbies[index] = value

    def add_hobby(self, hobby):
        self.hobbies.append(hobby)

person = Person("Dhanesh", 23)
person.add_hobby("Reading")
person.add_hobby("Cycling")

print(person)                # __str__
print(repr(person))          # __repr__
print(len(person))           # __len__
print(person[0])             # __getitem__
person[1] = "Swimming"       # __setitem__
print(person.hobbies)
```

```python
# operator overloading
''' 
__add__
__sub__
__mul__
__truediv__
__eq__
__lt__
__gt__
'''

class Vector:
    def __init__(self,x,y):
        self.x = x
        self.y = y
    
    def __add__(self,other):
        return Vector(self.x+other.x, self.y+other.y)
    
    def __sub__(self,other):
        return Vector(self.x-other.x,self.y-other.y)
    
    def __mul__(self,other):
        return Vector(self.x*other.x, self.y*other.y)
    
    def __str__(self):
        return f"({self.x},{self.y})"
    
v1 = Vector(2,3)
v2 = Vector(4,6)
print(v2-v1)
print(v1+v2)
print(v1-v2)
```
